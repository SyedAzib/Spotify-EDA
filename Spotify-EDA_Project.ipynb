```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('spotify_dataset.csv')
df
```

```python
# Checking is there any missing values
df.isnull().sum()
```

NOW WE MOVE TO EDA QUESTIONS


Q.01) How many unique artists are in the dataset?

```python
df['Artist'].nunique()
```

Q.02) What is the distribution of song popularity scores?

```python
df['Popularity'].describe()
```

```python
df['Popularity'].hist(bins=30,figsize=(10, 6))
plt.title('Distribution of Song Popularity')
plt.xlabel('Popularity')
plt.ylabel('Frequency')
plt.show()
```

Q.03) Which artists appear most frequently in the Top 200?

```python
top200 = df.sort_values(by='Popularity', ascending=False).head(200)
artist_counts = top200['Artist'].value_counts()
artist_counts
```

Q.04) What are the most common genres?

```python
common_genres = df['Genre'].value_counts()
common_genres
```

Q.05) What is the average duration of songs?

```python
df['Duration (ms)'] = pd.to_numeric(df['Duration (ms)'], errors='coerce')
df['Duration (ms)'].mean()
```

Q.06) How do audio features like danceability, energy, and valence vary across genres?

```python
# Convert the columns to numeric
df[['Danceability', 'Energy', 'Valence']] = df[['Danceability','Energy', 'Valence']].apply(pd.to_numeric, errors='coerce')
# Calculate the mean of the audio features by genre
avg_features = df.groupby('Genre')[['Danceability', 'Energy', 'Valence']].mean()
avg_features
```

VISUALIZATION

Q.01) Bar chart of Top 10 artists by number of chart appearances.

```python
top_artists = df['Artist'].value_counts().head(10)

plt.figure(figsize=(14, 8))
sns.barplot(x=top_artists.index, y=top_artists.values)
plt.title('Top 10 Artists by Number of Chart Appearances')
plt.xlabel('Artist')
plt.ylabel('Number of Chart Appearances')
plt.show()
```

Q.02) Histogram of song popularity scores.

```python
scores=df['Popularity'].value_counts()
plt.figure(figsize=(15, 8))
sns.histplot(df['Popularity'], bins=30, kde=True)
plt.title('Distribution of Song Popularity')
plt.xlabel('Popularity')
plt.ylabel('Frequency')
plt.show()
```

Q.03) Boxplot showing distribution of danceability across genres.

```python
plt.figure(figsize=(15, 8))
sns.boxplot(x='Genre', y='Danceability', data=df)
plt.title('Distribution of Danceability Across Genres')
plt.xticks(rotation=45)
plt.show()
```

Q.04) Heatmap showing correlation between audio features (danceability, energy, valence, tempo,
loudness).

```python
#The values of the upper mentioned datasets have got both strings and float
# As I have alreday converted some of it earlier
df[['Tempo', 'Loudness']] = df[['Tempo', 'Loudness']].apply(pd.to_numeric, errors='coerce')
```

```python
features = ['Danceability', 'Energy', 'Valence', 'Tempo', 'Loudness']
corr = df[features].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

Q.05) Line chart showing trend of popularity over time.

```python
# Convert 'Release Date' to datetime and extract the year
df['Release Date'] = pd.to_datetime(df['Release Date'], errors='coerce')
df['Year'] = df['Release Date'].dt.year

plt.figure(figsize=(15, 8))
sns.lineplot(x='Year', y='Popularity', data=df)
plt.title('Popularity Trend Over Time')
plt.xlabel('Year')
plt.ylabel('Popularity')
plt.show()
```

Q.06) Scatter plot of energy vs loudness colored by genre.

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x="Energy", y="Loudness", hue="main_Genre", alpha=0.7)

plt.title("Scatter Plot of Energy vs Loudness by Genre", fontsize=14)
plt.xlabel("Energy")
plt.ylabel("Loudness")
plt.legend(title="Genre", bbox_to_anchor=(1.05, 1), loc='upper left')  # move legend outside
plt.show()
```
